```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Indicadores-desempleo.csv')

#Renombramos las columnas para tener un mejor entendimiento de es cada una
df.columns = ['Periodos', 'Tasa_Desocupacion_Desestacionalizada', 'Tasa_Desocupacion_Tendencia_Ciclo']

#Verificar que los datos son correctos
print(df.head())
print(df.tail())
```

```python
# Convertimos los datos de la columna periodo a datetime para un mejor entendimiento de los datos
df['Periodos'] = pd.to_datetime(df['Periodos'], format='%Y/%m')

#Tamao de la grafica
plt.figure(figsize=(14, 7))

# Graficamos la primera columna
plt.plot(df['Periodos'], df['Tasa_Desocupacion_Desestacionalizada'], label='Desestacionalizada')

# Graficar la segunda columna
plt.plot(df['Periodos'], df['Tasa_Desocupacion_Tendencia_Ciclo'], label='Tendencia-Ciclo')

# Aadimos el titulo de la grafica y los nombres de los ejes
plt.title('Tasa de Desocupacin en Mxico')
plt.xlabel('Periodo')
plt.ylabel('Tasa de Desocupacin (%)')

# Aadir leyenda para saber que es cada linea
plt.legend()

# Mostramos la grafica
plt.show()
```
